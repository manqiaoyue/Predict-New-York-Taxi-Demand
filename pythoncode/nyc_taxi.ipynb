```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
os.chdir("/Users/shuozhang/Desktop/capstone/green taxi")
green = pd.read_csv("green_tripdata_2014-04.csv", index_col=False)
green.head()
```

```python
green.columns = [c.replace(" ", "") for c in green.columns] # Remove column name space 
green.drop([col for col in ["VendorID","Lpep_dropoff_datetime","Store_and_fwd_flag", "RateCodeID",\
                            "Dropoff_longitude","Dropoff_latitude","Passenger_count","Fare_amount",\
                           "Extra","MTA_tax","Ehail_fee","Trip_type","Tolls_amount"] \
                           if col in green], axis=1, inplace=True)
```

```python
green.head()
```

```python
green.columns=['Date', 'Lon', 'Lat', 'Distance', 'Tip', 'Total', 'Payment']
green["Type"] = "green"
green.head()
```

```python
green["Date"] =green["Date"].str[0:13].replace("\s+","-", regex=True)
green.head()
```

```python
################################################################################################################
## Uber Data
os.chdir("/Users/shuozhang/Desktop/capstone/uber")
uber = pd.read_csv("uber-raw-data-apr14.csv")
uber.head()
uber.columns=['Date', 'Lat', 'Lon', 'Base']
uber.drop(['Base'], axis=1, inplace=True)
uber["Type"] = "uber"
uber.head()
# print np.sum(uber.isnull())
```

```python
uber["Date"] = pd.to_datetime(uber["Date"]).dt.strftime('%Y-%m-%d-%H')
uber.head()
```

```python
################################################################################################################
## Weather
os.chdir("/Users/shuozhang/Desktop/capstone")
weather = pd.read_csv("weather.csv")
weather = pd.DataFrame(weather)
weather.columns = [c.replace(" ", "") for c in weather.columns]
weather = weather.drop_duplicates(subset='IDX', keep='first')
weather.head()
```

```python
weather["YR--MODAHRMN"] = pd.to_datetime(weather["YR--MODAHRMN"],format="%Y%m%d%H%M")
weather["YR--MODAHRMN"] = pd.to_datetime(weather["YR--MODAHRMN"]).dt.strftime('%Y-%m-%d-%H')
weather.drop([col for col in ["SPD","DEWP","IDX"] if col in weather], axis=1, inplace=True)
weather.columns=['Date', 'Temp', 'Rain', 'Snow']
```

```python
weather=weather.fillna(0)
weather.head()
np.sum(weather.isnull(), axis=0) 
```

```python
################################################################################################################
## Join Tables
```

```python
data=pd.concat([yellow, green, uber], axis=0)
```
