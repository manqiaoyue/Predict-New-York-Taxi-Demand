```python
import numpy as np
import pandas as pd
import os
os.chdir("/Users/shuozhang/Desktop/data")
xboost2=pd.read_csv('xboost2.txt', sep='|')
```

```python
xboost2.columns=map(lambda x: x.strip(), xboost2.columns)
```

```python
xboost2.columns
```

```python
xboost2['Value']=map(lambda x:abs(x), xboost2['Value'])
xboost2
```

```python
xboost2.dtypes
```

```python
XGB=xboost2.sort(['Value'], ascending=1)
```

```python
XGB.columns
```

```python
XGB.drop(['Step', ''], inplace=True, axis=1)
```

```python
XGB.index=np.arange(XGB.shape[0])
```

```python
XGB['n_estimators']=map(lambda x:int(x),XGB['n_estimators'] )
```

```python
XGB['max_depth']=map(lambda x:int(x),XGB['max_depth'] )
```

```python
import matplotlib.pyplot as plt
plt.subplot(221)
plt.plot(xboost2.colsample_bytree, xboost2.Value,label='colsample_bytree')
plt.plot(xboost2.gamma,xboost2.Value, label='gamma')
plt.plot(xboost2.subsample, xboost2.Value,label='subsample')
plt.legend()
plt.subplot(222)
plt.plot(xboost2.max_depth, xboost2.Value,label='max_depth')
plt.plot(xboost2.min_child_weight, xboost2.Value,label='min_child_weight')
plt.legend()
plt.subplot(223)
plt.plot(xboost2.n_estimators, xboost2.Value)
plt.title('n_estimators')
plt.subplot(224)
plt.plot(xboost2.learning_rate, xboost2.Value,)
plt.title('learning_rate')
plt.show()
```
